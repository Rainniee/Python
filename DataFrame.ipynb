# Intro to Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4
import warnings
warnings.filterwarnings('ignore')
```

```python
#Distribution
vis1 = sns.distplot(stats["InternetUsers"])
```

```python
#Boxplot
vis2 = sns.boxplot(data = stats, x='IncomeGroup', y='BirthRate')
```

```python
#Regression
vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', fit_reg=False, hue='IncomeGroup', size=8)
```

```python
#Keyword Arguments
vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', fit_reg=False, hue='IncomeGroup', size=8, scatter_kws={"s":100})
```
